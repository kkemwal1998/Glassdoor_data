```python
import pandas as pd
import numpy as np
import seaborn as se
```

Part-2

```python
Cleaned_data_P2 = pd.read_csv('Uncleaned_DS_jobs.csv')
Cleaned_data_P2
```

```python
#Calculating the company age
Curr_Year = 2023                   # Current Year
Cleaned_data_P2['Company Age'] = Curr_Year - Cleaned_data_P2['Founded']
Cleaned_data_P2
```

```python
#Removing coulmns named 'index','Founded',and 'competitors' from uncleaned dataset.
Columns_to_be_removed = ['index','Founded' ,'Competitors']
Cleaned_data_P2 = Cleaned_data_P2.drop(columns = Columns_to_be_removed,errors = 'ignore')
Cleaned_data_P2 
```

```python
#Removing glassdoor text from salary estimate column and K sign.
Cleaned_data_P2 ['Salary Estimate'] = Cleaned_data_P2 ['Salary Estimate'].str.replace(r'\s*\(Glassdoor est.\)\s*', '', regex=True)
Cleaned_data_P2 ['Salary Estimate'] = Cleaned_data_P2 ['Salary Estimate'].str.replace(r'[^0-9-]', '', regex=True)
Cleaned_data_P2 
```

```python
#Removing everything after n in company name
Cleaned_data_P2['Company Name'] = Cleaned_data_P2['Company Name'].str.split('\n', n=1, expand=True)[0]
Cleaned_data_P2
```

```python
#Extracting the state from the location 
Cleaned_data_P2['job_state'] = Cleaned_data_P2['Location'].str[-2:].str.strip()
Cleaned_data_P2
```

```python
# Remving -1 and replace it with na
Cleaned_data_P2 ['Industry'] = Cleaned_data_P2['Industry'].replace('-1','n/a',regex = True)
Cleaned_data_P2 ['Headquarters'] = Cleaned_data_P2['Headquarters'].replace('-1','n/a',regex = True)
Cleaned_data_P2 ['Size'] = Cleaned_data_P2['Size'].replace('-1','n/a',regex = True)

Cleaned_data_P2 ['Type of ownership'] = Cleaned_data_P2['Type of ownership'].replace('-1','n/a',regex = True)
Cleaned_data_P2
```

```python
# Splitting the salary into minium salary, average and maxiumum salary.
Cleaned_data_P2[['Min_salary','Max_salary']] = Cleaned_data_P2['Salary Estimate'].str.split('-',expand= True)
Cleaned_data_P2['Min_salary'] = pd.to_numeric(Cleaned_data_P2['Min_salary'])
Cleaned_data_P2['Max_salary'] = pd.to_numeric(Cleaned_data_P2['Max_salary'])
#Creating the average salary column
Cleaned_data_P2['AVG_salery'] = (Cleaned_data_P2['Min_salary']+Cleaned_data_P2['Max_salary'])/2
Cleaned_data_P2
```

```python
#Extracting the popular skills and making columns out of them.
############Step-8) Extracting polpular skills (Python, excel, tableau, power_bi, sql,aws) from job description.
#PYTHON
Cleaned_data_P2['PYTHON'] = Cleaned_data_P2['Job Description'].str.contains('Python',case = False)

Cleaned_data_P2['PYTHON'] = Cleaned_data_P2['PYTHON'].replace(True,1,regex=True)
Cleaned_data_P2['PYTHON'] = Cleaned_data_P2['PYTHON'].replace(False,0,regex=True)    
#Excel
Cleaned_data_P2['Excel'] = Cleaned_data_P2['Job Description'].str.contains('excel',case = False)
Cleaned_data_P2['Excel'] = Cleaned_data_P2['Excel'].replace(True,1,regex=True)
Cleaned_data_P2['Excel'] = Cleaned_data_P2['Excel'].replace(False,0,regex=True)    

#Tableau
Cleaned_data_P2['Tableau'] = Cleaned_data_P2['Job Description'].str.contains('tableau',case = False)
Cleaned_data_P2['Tableau'] = Cleaned_data_P2['Tableau'].replace(True,1,regex=True)
Cleaned_data_P2['Tableau'] = Cleaned_data_P2['Tableau'].replace(False,0,regex=True)    

#Power_bi
Cleaned_data_P2['Power bi'] = Cleaned_data_P2['Job Description'].str.contains('Power Bi',case = False)
Cleaned_data_P2['Power bi'] = Cleaned_data_P2['Power bi'].replace(True,1,regex = True)
Cleaned_data_P2['Power bi'] = Cleaned_data_P2['Power bi'].replace(False,0,regex = True)

#SQL
Cleaned_data_P2['SQL'] = Cleaned_data_P2['Job Description'].str.contains('sql',case = False)
Cleaned_data_P2['SQL'] = Cleaned_data_P2['SQL'].replace(True,1,regex=True)
Cleaned_data_P2['SQL'] = Cleaned_data_P2['SQL'].replace(False,0,regex=True)    

#AWS
Cleaned_data_P2['AWS'] = Cleaned_data_P2['Job Description'].str.contains('AWS',case = False)
Cleaned_data_P2['AWS'] = Cleaned_data_P2['AWS'].replace(True,1,regex=True)
Cleaned_data_P2['AWS'] = Cleaned_data_P2['AWS'].replace(False,0,regex=True)    
Cleaned_data_P2
```

```python
#Creating the job category column
def job_function(Title):
    if 'Data Scientist' in Title:
        return 'Data Scientist'
    elif 'ML' in Title or 'Machine Learning' in Title:
        return 'MLE'
    elif 'Business Intelligence Analyst I- Data Insights' in Title or 'Data Analyst II' in Title:
        return 'analyst'
    elif 'Data Engineer' in Title:
        return 'data engineer'
    elif 'Data Science Manager, Payment Acceptance - USA' in Title or 'Analytics Manager' in Title or 'Manager / Lead, Data Science & Analytics' in Title or 'Manager, Field Application Scientist, Southeast' in Title or 'Data Science Manager' in Title:
        return 'manager'
    elif 'Principal Scientist/Associate Director, Quality Control and Analytical Technologies' in Title or 'Director of Data Science' in Title or 'Principal Scientist/Associate Director, Quality Control and Analytical Technologies' in Title :
        return 'director'

    else:
        return 'NA'

Cleaned_data_P2['Job Category'] = Cleaned_data_P2['Job Title'].apply(job_function)
Cleaned_data_P2
```

```python
#############Step-10) Creating a column named 'Seniority'
Cleaned_data_P2['Seniority'] = Cleaned_data_P2['Job Title'].str.contains('Sr|Senior',case = False)
Cleaned_data_P2['Seniority'] = Cleaned_data_P2['Seniority'].replace(True,'senior')
Cleaned_data_P2['Seniority'] = Cleaned_data_P2['Seniority'].replace(False,'na')
Cleaned_data_P2
```

```python
# Remving everything in company'name after na
Cleaned_data_P2['Company Name'] = Cleaned_data_P2['Company Name'].str.split('\n', n=1, expand=True)[0]
Cleaned_data_P2
```

```python
# Therefore, we will be exporting our file to csv.
Cleaned_data_P2.to_csv('Cleaned_data_P2.csv')
```
